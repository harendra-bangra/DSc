```python
#importing the dataset
import pandas as pd
import numpy as np
ds = pd.read_csv(r"C:\Users\Yashpal\Desktop\3-2\dsc\lab\Datasets\iris.data",header=None)
ds
```

```python
#spliting the dataset in X and Y
X=ds.iloc[:,0:-1]
Y=ds.iloc[:,-1]
print("X is :\n",X)
print("\n\nY is :\n",Y)
```

```python
#scaling onto the feature vector
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X=sc.fit_transform(X)
X=np.around(X,2)
X
```

```python
#finding mean of each column
X_mean=np.mean(X,axis=0)
X_mean
```

```python
#substract mean for column from each entry
X=X-X_mean
X
```

```python
#covarience matrix
C=np.dot(X.T,X)
C
```

```python
#finding eigen values and eigen vectors
eival,eivec=np.linalg.eig(C)
print("Eigen values are :\n",eival)
print("\nEigen vectors are :\n",eivec)
```

```python
#sorting the eigen values in descending order and their eigen vectors
eival1=eival
eivec1=eivec
p=np.argsort(eival)
eival=[]
eivec=[]
p=np.flip(p)
print("P: ",p)
for i in p:
    eival.append(eival1[i])
    eivec.append(eivec1[i])
eival=np.array(eival)
eivec=np.array(eivec)
print("Eigen values are :\n",eival)
print("\nEigen vectors are :\n",eivec)
```

```python
#pov
pov=(eival/np.sum(eival))*100
pov
```

```python
#ploting
import matplotlib.pyplot as plt
comp=[1,2,3,4]
plt.plot(comp,pov)
plt.title('Percentage of variance vs Principal component')
plt.xlabel('Principal Component')
plt.ylabel('Percentage of variance')
plt.show()
```

```python
#consider only the features upto where the pov adds to <95% here 2
k=2
eisub=eivec[:,0:k]
eisub
```

```python
X_reduced=np.dot(eisub.T,X.T).T
X_reduced
```

```python
#seprating train and test dataset with test size 20%
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(X, Y, test_size=0.2, random_state=0)
print("X_Train dataset : \n",np.around(x_train,2))
print("Y_Train dataset : \n",y_train)
print("X_Test dataset : \n",np.around(x_test,2))
print("y_Test dataset : \n",y_test) 
```

```python
# Implement the logic of the algorithm using Gradient Descent Function
# Estimate linear regression coefficients using stochastic gradient descent
from math import exp
# Make a prediction with coefficients

def sigmoid(z):
  return 1.0 / (1.0 + np.exp(-z))

def Gradient_Descent(x_train, y_train, alpha, epochs):
  y_train = y_train.reshape(len(y_train), 1)
  coef = np.zeros([len(x_train[0])+1,1])
  x_train=np.insert(x_train,0,1,axis=1)
  # print(x_train)
  for e in range(epochs):
    cost = 0
    h_x = sigmoid(np.dot(x_train, coef))
    # cost = np.sum(-(y_train * np.log(h_x) + (1 - y_train) * np.log(1 - h_x)))
    # Updating theta
    coef[0] = coef[0] - alpha * np.sum(h_x - y_train)
    for i in range(1,len(coef)):
      coef[i] = coef[i] - alpha * np.sum((h_x - y_train) * x_train[:,i].reshape(-1,1))
  return coef

# Finding coefficients
coef = []
epochs = 500
alpha = 0.01
y_train_unique = set(y_train)
for i in y_train_unique:
  y_train_curr = np.where((y_train == i), 1, 0)
  coef.append(Gradient_Descent(x_train, y_train_curr, alpha, epochs))
print(coef)
```

```python
#predicting values
y_pred=[0 for i in range(len(y_test))]
h_x=[0 for i in range(len(y_test))]
x_test=np.insert(x_test,0,1,axis=1)
for i in range(len(coef)):
 h1=sigmoid(np.dot(x_test, coef[i]))
 for j in range(len(y_test)):
   if h_x[j]<h1[j]:
     h_x[j]=h1[j]
     y_pred[j]=i

print("Predicted values :\n",y_pred)
print("Actual values :\n",y_test)
```
